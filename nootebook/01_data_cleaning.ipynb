```python
clean_data = train_data.copy()
```

```python
#none_columns = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType']
#clean_data[none_columns] = clean_data[none_columns].fillna('None')
```

```python
target_columns = [
    "BsmtQual",
    "BsmtCond",
    "BsmtExposure",
    "BsmtFinType1",
    "BsmtFinType2"
]

clean_data[target_columns] = clean_data[target_columns].fillna("NoBasement")

target_columns = [
    "GarageType",
    "GarageFinish",
    "GarageQual",
    "GarageCond"
]

clean_data[target_columns] = clean_data[target_columns].fillna("NoGarage")

garage_numeric = [
    "GarageYrBlt",
    "GarageCars",
    "GarageArea"
]

clean_data[garage_numeric] = clean_data[garage_numeric].fillna(0)

clean_data["Fence"] = clean_data["Fence"].fillna("NoFence")
clean_data["FireplaceQu"] = clean_data["FireplaceQu"].fillna("NoFireplace")
clean_data["Alley"] = clean_data["Alley"].fillna("NoAlley")
clean_data["MasVnrArea"] = clean_data["MasVnrArea"].fillna(0)
clean_data["MasVnrType"] = clean_data["MasVnrType"].fillna("NoFeature")
clean_data["MiscFeature"] = clean_data["MiscFeature"].fillna("NoMansonry")
clean_data["LotFrontage"] = clean_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median())
)
clean_data["Electrical"] = clean_data["Electrical"].fillna(
    clean_data["Electrical"].mode()[0]
)
clean_data.isnull().sum().sort_values(ascending=False)
```
