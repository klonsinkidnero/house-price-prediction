```python
import pandas as pd
import numpy as np


feature_data = pd.read_csv('../data/clean_house_price.csv')

print(feature_data.shape)

print(feature_data["MiscFeature"].isnull().sum())

print(feature_data["MasVnrType"].isnull().sum())

feature_data[["MiscFeature","MasVnrType"]].head()
```

```python

feature_data['houseage'] = feature_data['YrSold'] - feature_data['YearBuilt']
feature_data['remoldedage'] = feature_data['YrSold'] - feature_data['YearRemodAdd']

feature_data["totalbathrooms"] = (
    feature_data["FullBath"] +
    0.5 * feature_data["HalfBath"] +
    feature_data["BsmtFullBath"] +
    0.5 * feature_data["BsmtHalfBath"]
)
feature_data[[
    "FullBath",
    "HalfBath",
    "BsmtFullBath",
    "BsmtHalfBath",
    "totalbathrooms"
]].head(10)
feature_data['totalporchsf'] = feature_data['OpenPorchSF'] + feature_data['EnclosedPorch'] + feature_data['3SsnPorch'] + feature_data['ScreenPorch']
feature_data['totalporchsf'].value_counts()
feature_data['totallivingarea'] = feature_data['1stFlrSF'] + feature_data['2ndFlrSF'] + feature_data['TotalBsmtSF']
feature_data['totallivingarea'].head(10)
feature_data['totaloutdoorspace'] = feature_data['WoodDeckSF'] + feature_data['OpenPorchSF'] + feature_data['EnclosedPorch']
feature_data['totaloutdoorspace'].head(10)
feature_data['haspool']= feature_data['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
feature_data['haspool'].value_counts()
feature_data['hasgarage']=feature_data['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
feature_data['hasgarage'].value_counts()
feature_data['hasbasement']=feature_data['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
feature_data['hasbasement'].value_counts()
feature_data["hasfireplace"] = feature_data['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
feature_data["hasfireplace"].value_counts()
feature_data["hasalley"] = feature_data['Alley'].apply(lambda x: 0 if pd.isna(x)  else 1)
feature_data['hasalley'].value_counts(dropna=False)
```

```python
feature_data.to_csv('../data/feature_house_price.csv',index=False)
```
